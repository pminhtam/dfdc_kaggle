```python
import random
from os import listdir
from os.path import isfile, join

import numpy as np

import json
import matplotlib.pyplot as plt
import cv2
from tqdm import tqdm
from mtcnn import MTCNN
import pickle
```

```python
detector = MTCNN()
```

```python
# data = json.load(open("/hdd/tam/kaggle/www.kaggle.com/c/16880/datadownload/dfdc_train_part_48/metadata.json"))
data = json.load(open("/data/tam/kaggle/train_sample_videos/metadata.json"))

margin = 0.2
```

```python
save_interval = 7 # perform face detection every {save_interval} frames

for vi in tqdm(data):
#     video = cv2.VideoCapture(join("/hdd/tam/kaggle/www.kaggle.com/c/16880/datadownload/dfdc_train_part_48/", vi))
    video = cv2.VideoCapture(join("/data/tam/kaggle/train_sample_videos/", vi))

    success = True
#     success, vframe = video.read()
#     save_interval = 5
#     if data[vi]['label'] == "FAKE":
# #         save_interval = 20
#         save_interval = 14
#     if data[vi]['label'] == 'REAL':
# #         save_interval = 5
#         save_interval = 3
    success, image = video.read()
    i = 0
    while success:
#         for i in range(0,video.__len__(),save_interval):
        for j in range(save_interval):
            success, image = video.read()
            if not success:
                break
#         print(image.shape)
#         if image.all() ==None:
#             continue
        try:
            image=cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
        except:
            continue
    #         face_positions = face_recognition.face_locations(img)
        face_positions = detector.detect_faces(image)
        if len(face_positions) == 0:
            continue
        face_position =face_positions[0]['box']
        x,y,w,h = face_position
        offsetx = round(margin * (w))
        offsety = round(margin * (h))
        y0 = round(max(y - offsety, 0))
        x1 = round(min(x + w + offsetx, image.shape[1]))
        y1 = round(min(y+ h + offsety, image.shape[0]))
        x0 = round(max(x - offsetx, 0))
        face = image[y0:y1,x0:x1]
        face = cv2.resize(face,(256,256))

        success, image = video.read()

#         if data[vi]['label'] == "FAKE":
#             plt.imsave(join("/hdd/tam/kaggle/train_imgs/df", vi+"_" + str(i)+".jpg"),face,format='jpg')
#         if data[vi]['label'] == 'REAL':
#             plt.imsave(join("/hdd/tam/kaggle/train_imgs/real", vi+ "_"+str(i)+".jpg"),face,format='jpg')
        if data[vi]['label'] == "FAKE":
            plt.imsave(join("/data/tam/kaggle/extract_raw_img_test/df", vi+"_" + str(i)+".jpg"),face,format='jpg')
        if data[vi]['label'] == 'REAL':
            plt.imsave(join("/data/tam/kaggle/extract_raw_img_test/real", vi+ "_"+str(i)+".jpg"),face,format='jpg')
        i+=1
```
