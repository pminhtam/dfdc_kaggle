```python
import pickle
import os
import glob
import numpy as np
import matplotlib.pyplot as plt
from os.path import isfile, join
from tqdm import tqdm
```

```python
data = glob.glob("/data/tam/kaggle/train_videos/real/*.pkl")
```

```python
for i in tqdm(data):
    k = 0
#     print(len(a))
    if os.path.exists(join("/hdd/tam/kaggle/train_imgs/real", i.split('/')[-1].split('.')[0] + "_" + str(k)+".jpg")):
        continue
#     print(len(a))
    a = pickle.load(open(i,'rb'))
    while k<len(a):
        plt.imsave(join("/hdd/tam/kaggle/train_imgs/real", i.split('/')[-1].split('.')[0] + "_" + str(k)+".jpg"),a[k],format='jpg')
        k+=7
```

```python
data_df = glob.glob("/hdd/tam/kaggle/train_videos/df/*.pkl")
for i in tqdm(data_df):
    k = 0
#     print(len(a))
    if os.path.exists(join("/hdd/tam/kaggle/train_imgs/df", i.split('/')[-1].split('.')[0] + "_" + str(k)+".jpg")):
        continue
    try:
        a = pickle.load(open(i,'rb'))
    except:
        continue
#     print(len(a))
    while k<len(a):
        plt.imsave(join("/hdd/tam/kaggle/train_imgs/df", i.split('/')[-1].split('.')[0] + "_" + str(k)+".jpg"),a[k],format='jpg')
        k+=7
```

```python
a = cv2.imread("/data/tam/kaggle/train_imgs/real/wynotylpnm.mp4_0.jpg")
```

```python
# plt.imshow(cv2.cvtColor(a,cv2.COLOR_BGR2RGB))
plt.imshow(cv2.cvtColor(a,cv2.COLOR_BGR2GAYSCALE))
```

```python
epsilon=1e-8
```

```python
f = np.fft.fft2(a)
fshift = np.fft.fftshift(f)
fshift += epsilon
magnitude_spectrum = np.log(np.abs(fshift))/12.0
```
